# Categorical 변수 데이터 탐색

```python
def display_boxplot(feature):
    with plt.style.context(('seaborn-dark')):
        sns.boxplot(x=df_x[feature], y=df_y['Price'], linewidth=2, palette="Set2")
```

```python
def display_multi_boxplots(start_feature_name, row, col):
    with plt.style.context(('seaborn-darkgrid')):
        f, axes = plt.subplots(row, col, figsize=(2.3*col, 3.5*row), sharey=True)

        nums = row*col
        start_idx = columns_cat.index(start_feature_name)

        for num, ax in zip(range(nums), axes.flat):
            try:
                sns.boxplot(x=df_x[columns_cat[start_idx + num]], y=df_y['Price'], palette="Set2", ax=ax)
            except:
                pass
```
